```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sampl = np.random.uniform(low=-2, high=2, size=(100000,))
```

```python
sampl
```

```python
def relu(data):
    relu_data=[]
    for i in range(0,len(data)):
        if data[i]<0:
            relu_data.append(0)
        else:
            relu_data.append(data[i])
    return relu_data
```

```python
data=relu(sampl)
```

```python
#relu
plt.figure(figsize=(6,5))
sns.lineplot(sampl,data)
plt.xlabel("z")
plt.ylabel("relu(z)")
plt.show()
```

```python
#leaky relu
def leaky_relu(data):
    leaky_relu_data=[]
    alpha=0.0001
    for i in range(0,len(data)):
        if data[i]<0:
            leaky_relu_data.append(alpha*data[i])
        else:
            leaky_relu_data.append(data[i])
    return leaky_relu_data
```

```python
leaky_data=leaky_relu(sampl)
```

```python
# leaky relu
plt.figure(figsize=(6,5))
sns.lineplot(sampl,leaky_data)
plt.xlabel("z")
plt.ylabel("leaky relu(z)")
plt.show()
```
